```python
import os
import pandas as pd
import pylab as pl
import numpy as np
%pylab inline
```

```python
#read citibike data
df = pd.read_csv('201508-citibike-tripdata.csv')
```

```python
#generate the age of male riders
df['agem'] = 2015-df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 1)]
```

```python
pl.figure(figsize = (15,15))
pl.plot(df['agem'],df['tripduration'],'o',color = '#53377A')
pl.xlabel('Age of male riders', fontsize = 15)
pl.ylabel('Tripduration', fontsize = 15)
pl.title('The Age of Male Citibike Riders and Their Ridership')
```
